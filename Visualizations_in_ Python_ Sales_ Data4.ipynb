# Visualizations in Python on Sales Data

### 1.Compare Sales by region for 2016 with 2015 using bar chart

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
import seaborn as sns
import os

# set the graphs to show in the jupyter notebook
%matplotlib inline

# set seabor graphs to a better style
sns.set(style="ticks")

SalesData = pd.read_csv('Downloads\Python case study 4\SalesData.csv' )
SalesData
```

### 1. Compare Sales by region for 2016 with 2015 using bar chart

```python
Region_Sales = SalesData.groupby("Region")["Sales2015","Sales2016"].sum()
plt.figure(figsize=(12,6))
Region_Sales.plot(kind="bar",figsize=(8,8))
plt.ylabel('Sales')
plt.legend(loc='upper left')
plt.title('SalesData by Regon')
plt.show()

```

### 2. What are the contributing factors to the sales for each region in 2016. Visualize it using a Pie Chart.

```python
sales_pie = SalesData.groupby("Region")["Sales2016"].sum()
```

```python
sales_pie
```

```python
plt.pie(sales_pie,autopct="%1.0f%%",labels=["Central","East","West"],shadow=True,explode=[0.0,0.1,0.0],colors=['r', 'g', 'b'])
plt.title("Contributing factors to the sales for each region in 2016")
plt.show()
```

### 3. Compare the total sales of 2015 and 2016 with respect to Region and Tiers 

```python
sales_region_tier = SalesData.groupby(["Region","Tier"])['Sales2015','Sales2016'].sum()
sales_region_tier
```

```python
sales_region_tier.plot(kind="bar")
plt.xlabel("Region/Tier")
plt.ylabel("Sales")
plt.title("Total sales of 2015 and 2016 with Region and Tiers")
plt.show()

```

### 6. Create a new column Qtr in the imported dataset. The Quarters are based on months and defined as -
â€¢ Jan - Mar : Q1
â€¢ Apr - Jun : Q2
â€¢ Jul - Sep : Q3
â€¢ Oct - Dec : Q4


```python
month =SalesData["Month"]
```

```python
quarter = []
for x in month :
    if x in ["Jan","Feb","Mar"]:
        quarter.append("Q1")
    elif x in ["Apr","May","Jun"]:
        quarter.append("Q2")
    elif x in ["Jul","Aug","Sep"]:
        quarter.append("Q3")
    else:
        quarter.append("Q4")
```

```python
quarter
```

```python
#create a new column "Qtr"
SalesData["Qtr"]= pd.Series(quarter)
SalesData
```

### 7. Compare Qtr wise sales in 2015 and 2016 in a bar plot 

```python
qtr_sales = SalesData.groupby("Qtr")["Sales2015","Sales2016"].sum()
qtr_sales 
```

```python
qtr_sales.plot(kind="bar")
plt.xlabel("Qtr")
plt.ylabel("Sales")
plt.title("Quarter wise sales in 2015 and 2016")
plt.show()
```

### 8. Determine the composition of Qtr wise sales in and 2016 with regards to all the Tiers in a pie chart

```python
qtr_pivot = SalesData.pivot_table(index='Qtr',columns='Tier',values='Sales2016')
qtr_pivot
```

```python
#Qtr Q1
plt.pie(x=qtr_pivot.loc["Q1",:],autopct="%1.0f%%",labels=["High","Low","Med","Out"],colors=['deepskyblue', 'darkorange', 'darkgreen'])
plt.show()
```

```python
#Qtr Q2
plt.pie(x=qtr_pivot.loc["Q2",:],autopct="%1.0f%%",labels=["High","Low","Med","Out"],colors=['deepskyblue', 'darkorange', 'darkgreen'])
plt.show()
```

```python
#Qtr Q3
plt.pie(x=qtr_pivot.loc["Q3",:],autopct="%-1.0f%%",labels=["High","Low","Med","Out"],colors=['deepskyblue', 'darkorange', 'darkgreen'])
plt.show()
```

```python
#Qtr Q4
plt.pie(x=qtr_pivot.loc["Q4",:],autopct="%1.0f%%",labels=["High","Low","Med","Out"],colors=['deepskyblue', 'darkorange', 'darkgreen'])
plt.show()                                                                                           
```
